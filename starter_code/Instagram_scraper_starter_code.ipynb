**Data Collection**

```python
#download likes, captions, comments

#dishy,feastbox, kurami, mindful chef, allplants, foodspring, hellofresh, simplycook0 gousto0
hashtags = ['gousto'] # , ['meat','vegeterian', 'vegetables', 'chicken', 'food', 'vegan', 'foodie']
posts = []

for h_tag in hashtags:

    hashtag = Hashtag.from_name(L.context, h_tag)

    SINCE = datetime(2020, 12, 9)
    UNTIL = datetime(2018, 12, 9)

    k = 0
    max = 500

    for post in takewhile(lambda p: p.date > UNTIL, dropwhile(lambda p: p.date > SINCE, hashtag.get_all_posts())):
        
        print(h_tag + ": " + str(k))
        post_arr = [post.caption, post.likes]
        comments_list = []
        comments_iter = post.get_comments()
        for comment in comments_iter:
            comments_list.append(comment.text)
        post_arr.append(comments_list)
        post_arr.append(h_tag)
        posts.append(post_arr)

        k +=1
        if k >=max:
            break
```

```python
print(len(posts))
```

```python
filtered_posts = []

for i in range(len(posts)):
    
    print(i)
    post = posts[i]
    comments = post[2]
    try:
        if detect(post[0]) != 'en':
            # check if caption is in english
            continue
    except (TypeError, LangDetectException):
        continue
    
    filtered_comments = []
    for comment in comments:
        try:
            if detect(comment) == 'en':
                filtered_comments.append(comment)
        except (TypeError, LangDetectException):
            continue
            
    filtered_post = [post[0], post[1], filtered_comments, post[3]]
    filtered_posts.append(filtered_post)
    
df = pd.DataFrame(filtered_posts, columns=['captions', 'likes', 'comments', 'hashtag'])
df.to_csv('gousto2018.csv', index=False)
```
